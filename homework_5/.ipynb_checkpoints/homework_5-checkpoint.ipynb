```python
for column in train.columns[1:]:
    delIndex = train.index[ (train[column] < test[column].min()) | (test[column].max() < train[column] )  ]
    train.drop( delIndex , inplace= True )
    
badCol = train.columns[[train[col].unique().size == 1 for col in train.columns ]]   
train.drop(columns = badCol , inplace = True )
test.drop(columns = badCol , inplace = True )

print('size :' + str(train.shape))
print('trget % :' + str(train['0'].value_counts(normalize =True)[1]))
train.head()
```

### Ensembling

```python
models = {'logReg' : logReg_best ,
          'baggingLogRer' : baggingLogRer_best ,
          'randomForest' :randomForest_best ,
          'extraTrees': extraTrees_best ,
          'decisionTree' : decisionTree_best ,}

prediction = pd.DataFrame( columns = [name for name in models] , index = X.index) 
prediction_proba = pd.DataFrame( columns = [name for name in models] , index = X.index) 

for trainVal , predVal in skf.split(X, y):
    for name in models:
        if name == 'logReg_best' or name == 'baggingLogRer_best' :
            models[name].fit( X.iloc[trainVal].drop(columns = corColumns)  , y.iloc[trainVal])
            prediction[name].iloc[predVal]  = models[name].predict( X.iloc[predVal].drop(columns = corColumns) )
            prediction_proba[name].iloc[predVal]  = pd.Series(models[name].predict_proba( X.iloc[predVal].drop(columns = corColumns))[:,1] , index= predVal)
        else:
            models[name].fit( X.iloc[trainVal] , y.iloc[trainVal])
            prediction[name].iloc[predVal] = models[name].predict(X.iloc[predVal])
            prediction_proba[name].iloc[predVal] =pd.Series( (models[name].predict_proba(X.iloc[predVal]))[:,1] , index= predVal)
```
